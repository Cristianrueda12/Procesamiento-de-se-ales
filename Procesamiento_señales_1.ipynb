```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
frecuencia = 50  # Frecuencia de la señal en Hz
duracion = 1  # Duración en segundos
frecuencia_muestreo_alta = 1000  # Frecuencia de muestreo alta en Hz
frecuencia_muestreo_baja = 150  # Frecuencia de muestreo baja en Hz

# Generar el tiempo y las señales
t_alta = np.linspace(0, duracion, int(frecuencia_muestreo_alta * duracion), endpoint=False)
t_baja = np.linspace(0, duracion, int(frecuencia_muestreo_baja * duracion), endpoint=False)

senal_alta = np.sin(2 * np.pi * frecuencia * t_alta)
senal_baja = np.sin(2 * np.pi * frecuencia * t_baja)
```

```python
# Graficar la señal con frecuencia de muestreo alta
plt.figure(figsize=(10, 4))
plt.plot(t_alta, senal_alta, 'b-', label='Frecuencia de muestreo alta ({} Hz)'.format(frecuencia_muestreo_alta))
plt.title('Señal con frecuencia de muestreo alta')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
# Graficar la señal con frecuencia de muestreo baja
plt.figure(figsize=(10, 4))
plt.plot(t_baja, senal_baja, 'r-', label='Frecuencia de muestreo baja ({} Hz)'.format(frecuencia_muestreo_baja))
plt.title('Señal con frecuencia de muestreo baja')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
# Comparar ambas señales en una sola gráfica
plt.figure(figsize=(10, 6))
plt.plot(t_alta, senal_alta, 'b-', label='Frecuencia de muestreo alta ({} Hz)'.format(frecuencia_muestreo_alta))
plt.plot(t_baja, senal_baja, 'r-', label='Frecuencia de muestreo baja ({} Hz)'.format(frecuencia_muestreo_baja))
plt.title('Comparación de señales con diferentes frecuencias de muestreo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
def calcular_fft(senal, frecuencia_muestreo):
    # Calcular la FFT
    N = len(senal)  # Número de puntos
    fft = np.fft.fft(senal)  # Transformada rápida de Fourier
    fft = np.abs(fft[:N // 2])  # Tomar la magnitud y la mitad (frecuencias positivas)
    frecuencias = np.fft.fftfreq(N, 1 / frecuencia_muestreo)[:N // 2]  # Eje de frecuencias
    return frecuencias, fft
```

```python
# FFT para la señal con frecuencia de muestreo alta
frecuencias_alta, fft_alta = calcular_fft(senal_alta, frecuencia_muestreo_alta)

# FFT para la señal con frecuencia de muestreo baja
frecuencias_baja, fft_baja = calcular_fft(senal_baja, frecuencia_muestreo_baja)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(frecuencias_alta, fft_alta, 'b-', label='Frecuencia de muestreo alta ({} Hz)'.format(frecuencia_muestreo_alta))
plt.title('Espectro de frecuencia - Señal con frecuencia de muestreo alta')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(frecuencias_baja, fft_baja, 'r-', label='Frecuencia de muestreo baja ({} Hz)'.format(frecuencia_muestreo_baja))
plt.title('Espectro de frecuencia - Señal con frecuencia de muestreo baja')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.legend()
plt.grid()
plt.show()
```
